## Create `name_2` from col `name`

First, we will look at columns `product_full_name` and `name`. Since `product_full_name` seems to contain more info in its column, we decide to take any value from `product_full_name` and fill in column `name` if any cell in `name` happen to be `NaN`


```python
# fill in NaN value with product_full_name value

# define a function for 2 columns product_full_name, name

def test(product_full_name, product_full_name_len, name, name_len):
    if name_len == 0 :
        return product_full_name
    else:
        return name

# create a col name_2 that displays name value
# if name_len = zero, then display value in product_full_name, otherwise, use value in name column

df234['name_2'] = df234.apply(lambda x: test(x['product_full_name'],x['product_full_name_len'],x['name'], x['name_len'] ), axis=1)

# #check 
# df234.loc[df234['brand_name_len'] != df234['name_len']]

#drop unused col & retain columns to retain 'name_2'
df234 = df234.drop(['name','name_len'], axis=1) 

df234
```

## combine `brand_description` and `description`

We will create a new column called `branddescription_description_combine` which contain the value of both brand_description and description with below conditions

If the lengths are the same, in each row, we can display either value from brand_description/ description

If the length are different, we will combine those into 1 value like what we did earlier for `product_full_name` and `name_2`

However, if the length difference is less than 5, we can still assume those are basically the same context since there may be there different in typing/ typo/ plurals etc. those shouldn't be a huge effect on the  meanning of the whole sentence


```python
# define a function for 2 columns brand_description,description

def test(brand_description, brand_description_len, description, description_len, len_x_y):
    if brand_description_len == description_len :
        return brand_description # if brand_name and name col have same value, display either one (in this case, use brand_description value)
    elif abs(len_x_y) <= 5:
        return brand_description # if the difference in length is less than 5 characters, we can still take it as the same value and display either one
    else:
        return brand_description, description # if brand_name and name col have different value, then display both in 1 cell

# create a col brandname_name_combine that displays brand_name/name value
df234['branddescription_description_combine'] = df234.apply(lambda x: test(x['brand_description'], x['brand_description_len'],x['description'], x['description_len'], x['len_x_y'] ), axis=1)
# check
df234.loc[df234['brand_description_len'] != df234['description_len']]
```

# Top 50 brands
   
   Group by the top 50 brands that had the most products for sale. 

   Assign an ordered identifier number (0 - 49) for the top 50 brands and the identifier 99 for all other brands. 
   
   NOTE: Order of number assignment does not imply an ordinal importance.

```python
## make brands all lower case 

df234['brand'] = df234['brand'].str.lower()
df234['branddescription_description_combine'] = df234['branddescription_description_combine'].str.lower()
```

 Create a sub-DataFrame as a group by on brands with a count of their available number of products for sale, sorting on the top 50 brands that had the most products for sale. An ordinal number is assigned to our top 50 brands in the section below:

```python
# group by brand and count their number of products, sorting from the highest to the lowest amount of products
  # save top 50 
  
top50 = df234.groupby('brand')['product_id'].count()

top50 = top50.reset_index()  # save groupby as complete dataframe with both columns

top50 = top50.sort_values(by='product_id', ascending=False)

top50.head(50)
# save list of top 50 brands

top50 = top50.iloc[0:50]

top50['brand_number'] = top50['brand'].factorize()[0]

top50
```

Save brand numbers to a dictionary and map them back to our main DataFrame.
We assign 99 as the number identifier to all other brands

### Assign number to top 50 brands by `product_id` count


```python
## Assign numbers for top 50 brands based on product_id
    
top50dict = pd.Series(top50.brand_number.values, index=top50.brand).to_dict()

## Assign number to each of the top 50 brands

df234['brand_number'] = df234['brand']

df234['brand_number'] = df234['brand_number'].map(top50dict)

## Assign number to all other brands
df234['brand_number'].fillna(99, inplace=True)
```

# TF-IDF Vectorizer  Logistic Regression Model
  The TF-IDF vectorizer will assign a number vector to each product based on the frequency unique words used in the products description field, relative to presence of said words over all of the documents in our word corpus. Our word corpus comprised of roughly 268,000 product descriptions

Use sklearn TF-IDF and stopwords package to initiate a TF-IDF vectorizer.
Our vectorizer will look for naturally ocurring unigram through trigram occurrences of words. 

Parameters include: unigram through tri-gram expressions, minimum word occurrences set to 5 times in the corpus; 5,000 total features, and remove English stopwords from the corpus

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords

vectorizer = TfidfVectorizer(ngram_range=(1,3),
                             max_features = 5000,
                             min_df=5,
                             token_pattern=r'\b[a-zA-Z]{3,}\b',
                             max_df=0.4, stop_words=stopwords.words('english'))
```

Save combined-product-description column as our corpus, with all of string values within saved as a list.

Transform our corpus into a TF-IDF representation using the parameters established when creating our vectorizer (above).

```python
# checking NaN values
df234["branddescription_description_combine"].isnull().sum()
```

```python
# fill in NaN values
df234["branddescription_description_combine"]=df234["branddescription_description_combine"].fillna("")
```

```python
## perform TFIDF Vectorizer

corpus = list(df234["branddescription_description_combine"].values)

X = vectorizer.fit_transform(corpus)
```

Use built in logistic regression in sklearn to train and predict which the brand using our TF-IDF vecorizer.

Perform a training / test data split by a ratio of 80/20 respectively, while stratifying our two data sets to ensure equal representation of all of the brands (top 50 + other brands).

```python
from sklearn.linear_model import LogisticRegression


X   #<-- X takes the tf-idf vector as the values to regress on 
y = df234["brand_number"].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,
                                                    random_state = 0, stratify=y)

## Initiate and fit model
lr = LogisticRegression(max_iter=300)
lr.fit(X, y)

## Accuracy on baseline logistic regression

y_pred = lr.predict(X)

# calculate accuracy
np.mean(y_pred == y)
```

Our Logistic Regression model produced a 86.67% accuracy rate.
